```python
import skimage.io as io
import nibabel as nib
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
```

```python
#Masque initial
path=r'C:\Users\zelmouaffek\Desktop\Prétraitement\Mask.jpeg'
I=io.imread(path)
plt.imshow(I,cmap='gray')
```

## Binarisation par Otsu

```python
seuil=threshold_otsu(I) # Otsu trouve le seuil
J=I>seuil # on garde que les pixels suppérieur au seuil ( PixelS grix deviennent=1, donc blanc )
plt.imshow(J,cmap='gray')
io.imsave('new_mask.jpeg',J.astype('uint8'))
```
